
## Problem Statement
Udemy hosts thousands of courses across diverse subjects, but not all courses attract the same level of interest or engagement. Course creators and platform designers often lack clarity about which factors influence student enrollment, pricing success, and content effectiveness. Without this understanding, it becomes difficult to design courses that align with learner preferences and market demand.

This project seeks to address the problem by analyzing a comprehensive Udemy course dataset to identify:

What characteristics correlate with higher subscriber counts

How pricing and course level influence student engagement

How subjects differ in popularity

Whether course structure (lectures, duration) affects reviews or enrollment

Through systematic EDA, the goal is to uncover meaningful insights that can guide educational content creators and improve learner experience on Udemy.
